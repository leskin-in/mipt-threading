# Метод Монте-Карло для модели случайных блужданий: результаты

```python
import matplotlib.pyplot as plt
```

Импортируем и обработаем исходные данные:

```python
DATA_1_SIZE = 5
DATA_2_SIZE = 5
```

```python
data_1 = []
data_2 = []
data_0 = []

counter = 0
with open("stats.txt", "r") as f:
    for line in f:
        counter += 1
        if ((counter == DATA_1_SIZE + 1) or (counter == DATA_1_SIZE + 1 + DATA_2_SIZE + 1)):
            continue
        temp_list = line.split(" ")
        temp_list[2] = temp_list[2][:-1]
        if (counter <= DATA_1_SIZE):
            data_1.append(temp_list)
        elif (counter <= DATA_1_SIZE + 1 + DATA_2_SIZE):
            data_2.append(temp_list)
        else:
            data_0.append(temp_list)

data_0_T = [float(e[2][:-1]) for e in data_0]

data_1_N = [int(data_1[i][6]) for i in range(DATA_1_SIZE)]
data_1_T = [float(data_1[i][2]) for i in range(DATA_1_SIZE)]
data_1_S = [data_0_T[i] / data_1_T[i] for i in range(DATA_1_SIZE)]
data_1_E = [data_1_S[i] / float(data_1[i][8]) for i in range(DATA_1_SIZE)]

data_2_P = [int(data_2[i][8]) for i in range(DATA_2_SIZE)]
data_2_T = [float(data_2[i][2]) for i in range(DATA_2_SIZE)]
data_2_S = [data_0_T[DATA_1_SIZE] / data_2_T[i] for i in range(DATA_2_SIZE)]
data_2_E = [data_2_S[i] / float(data_2[i][8]) for i in range(DATA_2_SIZE)]
```

Построим графики для первого набора данных (зависимости от $N$):

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$N$ (число опытов)', fontsize=16)
plt.ylabel(r'$T$ (время исполнения цикла)', fontsize=16)
plt.plot(data_1_N, data_1_T, color="#9f8170", marker="8")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$N$ (число опытов)', fontsize=16)
plt.ylabel(r'$S$ (ускорение по отношению к однопроцессорному исполнению)', fontsize=16)
plt.plot(data_1_N, data_1_S, color="#44944a", marker="8")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$N$ (число опытов)', fontsize=16)
plt.ylabel(r'$E$ (эффективность распараллеливания в пересчёте на один процесс)', fontsize=16)
plt.plot(data_1_N, data_1_E, color="#f54021", marker="8")
plt.show()
```

Построим графики для второго набора данных (зависимости от $N$):

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$P$ (число процессов)', fontsize=16)
plt.ylabel(r'$T$ (время исполнения цикла)', fontsize=16)
plt.plot(data_2_P, data_2_T, color="#9f8170", marker="8")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$P$ (число процессов)', fontsize=16)
plt.ylabel(r'$S$ (ускорение по отношению к однопроцессорному исполнению)', fontsize=16)
plt.plot(data_2_P, data_2_S, color="#44944a", marker="8")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$P$ (число процессов)', fontsize=16)
plt.ylabel(r'$E$ (эффективность распараллеливания в пересчёте на один процесс)', fontsize=16)
plt.plot(data_2_P, data_2_E, color="#f54021", marker="8")
plt.show()
```
