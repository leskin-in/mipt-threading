# Параллельная сортировка слиянием: результаты

```python
import matplotlib.pyplot as plt
```

Импортируем и обработаем исходные данные:

```python
DATA_SIZE = 5
```

```python
data_1 = []
data_q = []

counter = 0
with open("stats.txt", "r") as f:
    for line in f:
        counter += 1
        temp_list = line.split(" ")
        temp_list[0] = temp_list[0][:-1]
        if (counter <= DATA_SIZE):
            data_1.append(temp_list)
        else:
            data_q.append(temp_list)

data_1_P = [int(data_1[i][3]) for i in range(DATA_SIZE)]
data_1_T = [float(data_1[i][0]) for i in range(DATA_SIZE)]
data_1_S = [data_1_T[0] / data_1_T[i] for i in range(DATA_SIZE)]
data_1_E = [data_1_S[i] / float(data_1_P[i]) for i in range(DATA_SIZE)]
```

Построим графики зависимости параметров работы программы от $P$:

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$P$ (число процессов)', fontsize=16)
plt.ylabel(r'$T$ (время исполнения сортировки)', fontsize=16)
plt.plot(data_1_P, data_1_T, color="#9f8170", marker="8", label="mergesort")
plt.plot(data_1_P, [data_q[0][0] for i in range(len(data_1_P))], color="#7e0059", marker="", label="qsort")
plt.legend(loc = 1, prop = {"size": 16})
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$P$ (число процессов)', fontsize=16)
plt.ylabel(r'$S$ (ускорение по отношению к однопроцессорному исполнению)', fontsize=16)
plt.plot(data_1_P, data_1_S, color="#44944a", marker="8")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel(r'$P$ (число процессов)', fontsize=16)
plt.ylabel(r'$E$ (эффективность распараллеливания в пересчёте на один процесс)', fontsize=16)
plt.plot(data_1_P, data_1_E, color="#f54021", marker="8")
plt.show()
```
